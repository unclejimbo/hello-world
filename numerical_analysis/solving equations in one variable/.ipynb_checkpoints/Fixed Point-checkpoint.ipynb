```julia
using PyPlot

function fixed_point(f, p0, tolerance, max_iteration)
    eq_x = linspace(1.25, 1.55, 100) 
    eq_y = zeros(0)
    for x in eq_x
        push!(eq_y, f(x))
    end
    plot(eq_x, eq_y)
    plot(eq_x, eq_x)
    
    x_vec = zeros(0)
    y_vec = zeros(0)
    i = 1
    
    while i <= max_iteration
        println("p is $p0")
        p = f(p0)
        plot(p0, p, "ro")
        plot([p0, p], [p, p])
        plot([p, p], [p, f(p)])
        if abs(p - p0) < tolerance
            println("Root of $f is $p.")
            println("Iteration loops: $i.")
            return p
        end
        i += 1
        p0 = p
    end
    
    println("Algorithm did not converge.")
    return
end
```

```julia
f1(x) = x-x^3-4x^2+10
fixed_point(f1, 1.5, 0.0001, 10)
```

```julia
f2(x) = sqrt(10/x - 4x)
fixed_point(f2, 1.5, 0.0001, 10)
```

```julia
f3(x) = sqrt(10-x^3)/2
fixed_point(f3, 1.5, 0.0001, 30)
```
