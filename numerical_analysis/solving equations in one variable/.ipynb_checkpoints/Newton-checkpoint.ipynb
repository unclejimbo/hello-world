```julia
using PyPlot

function Newton(f, df, p0, tolerance, max_iteration)
    x = linspace(0, 2, 100)
    y = zeros(0)
    for dx in x
        push!(y, f(dx))
    end
    plot(x, y)
    
    i = 1
    while i <= max_iteration
        p = p0 - f(p0)/df(p0)
        plot(p, f(p), "ro")

        if abs(p-p0) < tolerance
            println("Root of $f is $p.")
            return p
        end
        i += 1
        p0 = p
    end
    
    println("Algorithm did not converge.")
    return 
end
```

```julia
f1(x) = x^2-2x+1
df1(x) = 2x-2
Newton(f1, df1, 0.5, 0.0001, 20)
```
